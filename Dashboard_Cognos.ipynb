```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Load the dataset with error handling
tech_url = "/content/m5_survey_data_technologies_normalised.csv"
demo_url = "/content/m5_survey_data_demographics.csv"

df_tech = pd.read_csv(tech_url)

try:
    df_demo = pd.read_csv(demo_url, on_bad_lines='skip')
except pd.errors.ParserError as e:
    print(f"Error parsing the demographics file: {e}")

# Set up plotting aesthetics
sns.set(style="whitegrid")
```

# **Current Technology Usage Tab**


```python
# Top 10 LanguageWorkedWith
languages = df_tech['LanguageWorkedWith'].str.split(';', expand=True).stack().value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=languages.index, x=languages.values, palette="viridis")
plt.title('Top 10 Languages')
plt.xlabel('Count')
plt.ylabel('Languages')
plt.show()


```

```python
# Top 10 DatabaseWorkedWith
databases = df_tech['DatabaseWorkedWith'].str.split(';', expand=True).stack().value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=databases.index, x=databases.values, palette="viridis")
plt.title('Top 10 Databases')
plt.xlabel('Count')
plt.ylabel('Databases')
plt.show()

```

```python
# PlatformWorkedWith
platforms = df_tech['PlatformWorkedWith'].str.split(';', expand=True).stack().value_counts()
platforms_wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(platforms)
plt.figure(figsize=(10, 6))
plt.imshow(platforms_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Platforms')
plt.show()

```

```python
# Top 10 WebFrameWorkedWith
webframes = df_tech['WebFrameWorkedWith'].str.split(';', expand=True).stack().value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.scatterplot(y=webframes.index, x=webframes.values, size=webframes.values, sizes=(100, 1000), legend=False, palette="viridis")
plt.title('Top 10 Web Frameworks')
plt.xlabel('Count')
plt.ylabel('Web Frameworks')
plt.show()
```

# **Demographics Tab**

```python
# Filter Gender data
df_demo = df_demo[df_demo['Gender'].isin(['Man', 'Woman'])]

# Respondent classified by gender
gender_count = df_demo['Gender'].value_counts()
plt.figure(figsize=(10, 6))
gender_count.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("viridis", len(gender_count)))
plt.title('Respondent Count by Gender')
plt.ylabel('')
plt.show()

```

```python
# Respondent count for countries
country_count = df_demo['Country'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=country_count.index, x=country_count.values, palette="viridis")
plt.title('Respondent Count for Countries')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()


```

```python
# Respondent count by age
age_count = df_demo['Age'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x=age_count.index, y=age_count.values, marker='o', palette="viridis")
plt.title('Respondent Count by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

```

```python
# Respondent count by gender and classified by education level
gender_education = df_demo.groupby(['Gender', 'EdLevel']).size().unstack().fillna(0)
plt.figure(figsize=(14, 8))
gender_education.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Respondent Count by Gender and Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import squarify  # for hierarchy bubble chart

# Set up plotting aesthetics
sns.set(style="whitegrid")

# Create a figure with 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(20, 15))

# Panel 1: Top 10 LanguageWorkedWith
languages = df_tech['LanguageWorkedWith'].str.split(';', expand=True).stack().value_counts().head(10)
sns.barplot(ax=axs[0, 0], y=languages.index, x=languages.values, palette="viridis")
axs[0, 0].set_title('Top 10 Languages')
axs[0, 0].set_xlabel('Count')
axs[0, 0].set_ylabel('Languages')
for i, v in enumerate(languages.values):
    axs[0, 0].text(v + 1, i, str(v), color='black', va='center')

# Panel 2: Top 10 DatabaseWorkedWith
databases = df_tech['DatabaseWorkedWith'].str.split(';', expand=True).stack().value_counts().head(10)
sns.barplot(ax=axs[0, 1], y=databases.index, x=databases.values, palette="viridis")
axs[0, 1].set_title('Top 10 Databases')
axs[0, 1].set_xlabel('Count')
axs[0, 1].set_ylabel('Databases')
for i, v in enumerate(databases.values):
    axs[0, 1].text(v + 1, i, str(v), color='black', va='center')

# Panel 3: PlatformWorkedWith
platforms = df_tech['PlatformWorkedWith'].str.split(';', expand=True).stack().value_counts()
platforms_wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(platforms)
axs[1, 0].imshow(platforms_wordcloud, interpolation='bilinear')
axs[1, 0].axis('off')
axs[1, 0].set_title('Platforms')

# Panel 4: Top 10 WebFrameWorkedWith
webframes = df_tech['WebFrameWorkedWith'].str.split(';', expand=True).stack().value_counts().head(10)
sizes = webframes.values
labels = webframes.index
colors = sns.color_palette("viridis", len(sizes))
axs[1, 1].barh(labels, sizes, color=colors)
axs[1, 1].set_title('Top 10 Web Frameworks')
axs[1, 1].set_xlabel('Count')
axs[1, 1].set_ylabel('Web Frameworks')

# Adjust layout and display
plt.tight_layout()
plt.show()
```
